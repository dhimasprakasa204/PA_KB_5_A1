# Import Library

```python
import numpy as np
import pandas as pd 
import matplotlib.pyplot as plt
import tensorflow as tf
import keras, os, cv2
import splitfolders
from keras_preprocessing.image import ImageDataGenerator
from keras.models import Sequential, Model
from keras.layers import Dense, Dropout, Flatten, Conv2D, MaxPooling2D, Concatenate, Input
from tensorflow.keras import layers, optimizers, datasets
from tensorflow.keras.optimizers import SGD
from sklearn import preprocessing
from pathlib import Path
from skimage.io import imread, imshow
from skimage.color import rgb2gray
from skimage.transform import rescale
```

# Split Data Folder

```python
train_path = r'./../datasets/train_folder'
test_path = r'./../datasets/test_folder'
val_path = r'./../datasets/PA/val'
```

### Split Data dengan Ratio 70% train, 20% Validation, 10% Test

```python
splitfolders.ratio(
    train_path,
    output='PA',
    ratio=(.70,.20,.10)
)
```

```python
splitfolders.ratio(
    test_path,
    output='PA',
    ratio=(.70,.20,.10)
)
```

# Data Collecting

### Menampilkan Banyak Data Train, Test dan Validation

```python
train_datagen = ImageDataGenerator(
    rescale=1./255, 
    horizontal_flip = True,
    rotation_range= 30,
    shear_range=.2,
    zoom_range=.2,
    fill_mode='nearest'
    
)

test_datagen = ImageDataGenerator(
    rescale=1./255, 
    horizontal_flip = True,
    rotation_range= 30,
    shear_range=.2,
    zoom_range=.2,
    fill_mode='nearest'
    
)

val_datagen = ImageDataGenerator(
    rescale=1./255, 
    
)
```

```python
print('Traning Generator: ', end="")
train_generator = train_datagen.flow_from_directory(
    train_path,
    target_size = (150,150),
    shuffle=True,
    batch_size = 32,
    class_mode ='binary'
    
)

print('Testing Generator: ', end="")
test_generator = test_datagen.flow_from_directory(
    test_path,
    target_size = (150,150),
    shuffle=True,
    batch_size = 32,
    class_mode ='binary'
    
)

print('Validation Generator: ', end="")
val_generator = val_datagen.flow_from_directory(
    val_path,
    target_size = (150,150),
    shuffle=True, 
    batch_size = 32,
    class_mode ='binary'
    
)
```

# Data Modelling

```python
model = Sequential([
    Conv2D(32, (3,3), activation='relu', input_shape=(150,150, 3)),
    MaxPooling2D(2,2),
    Conv2D(64, (3,3), activation='relu'),
    MaxPooling2D(2,2),
    Flatten(),
    Dense(1, activation='sigmoid')

])
```

```python
model.compile(
    optimizer=tf.keras.optimizers.Adam(),
    loss=tf.keras.losses.binary_crossentropy,
    metrics=['accuracy']
)
```

```python
history = model.fit(
    train_generator,
    epochs=20,
    validation_data = val_generator
)
```

## Visualisasi Akurasi

```python
#train
y_acc_train = history.history['accuracy']
y_acc_val = history.history['val_accuracy']

#validation
y_loss_train = history.history['loss']
y_loss_val = history.history['val_loss']

#epoch
x_ = range(len(y_acc_train))
```

```python
plt.figure(1, figsize=(5,4))
plt.plot(x_, y_acc_train, c='r', label='Training')
plt.plot(x_, y_acc_val, c='b', label='Validation')
plt.title('Akurasi')
plt.xlabel('Epochs')
plt.ylabel('Akurasi')
plt.legend()

plt.figure(2, figsize=(5,4))
plt.plot(x_[3:], y_loss_train[3:], c='r', label='Training')
plt.plot(x_[3:], y_loss_val[3:], c='b', label='Validation')
plt.title('Loss')
plt.xlabel('Epochs')
plt.ylabel('Loss')
plt.legend()
plt.show()
```

```python
model.evaluate(val_generator)
```
